```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def stephan_boltzman(R,T):
    sigma =5.67e-8
    L=4*np.pi*R**2*sigma*T**4
    return L
```

```python
star_names = ["Sirius", "Canopus", "Rigil Kentaurus", "Arcturus", "Vega", "Capella", "Rigel", "Procyon", "Achernar", 
"Betelgeuse", "Acrux", "Altair", "Aldebaran", "Spica", "Antares", "Pollux", "Fomalhaut", "Deneb", "Regulus", "Adhara"]

star_temperatures =  np.array([9940, 7400, 5800, 4300, 9600, 4900, 12100, 6550, 14600, 3500, 28000, 7700, 3900, 22200, 3400, 4940, 8550, 8525, 12300, 20800])

star_radii = np.array([1.711e9, 7.8e8, 1.227e9, 2.785e9, 2.364e9, 1.192e9, 7.6e8, 7.43e8, 1.586e9, 9.52e10, 6.01e9, 1.656e9, 
4.28e9, 7.739e8, 8.12e9, 1.927e9, 1.834e9, 8.48e9, 3.919e9, 4.66e9])
```

```python
Lumpem=np.abs(stephan_boltzman(star_radii,star_temperatures/10))
```

```python
plt.scatter(star_temperatures, Lumpem)
plt.xlabel('Temperatura K')
plt.ylabel("Luminosidad")
plt.xscale("log")
plt.yscale("log")
plt.show()
```
